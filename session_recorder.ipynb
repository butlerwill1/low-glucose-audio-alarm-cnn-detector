# Audio Session Recorder

This notebook records audio sessions for machine learning training data collection.

**Purpose**: Collect raw audio data for glucose alarm classification

**Output**: WAV files (16kHz, mono, 16-bit PCM)

---

## Instructions

1. Set the session parameters in the configuration cell below
2. Run all cells
3. The recording will start automatically
4. Files are saved to the `sessions/` directory

```python
# Imports
import sounddevice as sd
import numpy as np
from scipy.io import wavfile
import os
from datetime import datetime
```

## Configuration

**Modify these variables before each recording session:**

```python
# ============================================================================
# SESSION CONFIGURATION - MODIFY THESE VALUES
# ============================================================================

# Session type: "glucose_alarm" or "no_glucose_alarm"
SESSION_TYPE = "glucose_alarm"

# Background noise: "background_noise" or "no_background_noise"
BACKGROUND_NOISE = "no_background_noise"

# Recording duration in seconds (60-300 recommended)
DURATION_SECONDS = 180

# ============================================================================
# AUDIO SETTINGS - DO NOT MODIFY UNLESS NEEDED
# ============================================================================

SAMPLE_RATE = 16000  # Hz
CHANNELS = 1         # Mono
DTYPE = 'int16'      # 16-bit PCM
OUTPUT_DIR = 'sessions'
```

## Recording Function

```python
def record_session(session_type, background_noise, duration_seconds, sample_rate=SAMPLE_RATE, 
                   channels=CHANNELS, output_dir=OUTPUT_DIR):
    """
    Record an audio session and save it with metadata in the filename.
    
    Parameters:
    -----------
    session_type : str
        Type of session: "glucose_alarm" or "no_glucose_alarm"
    background_noise : str
        Background noise level: "background_noise" or "no_background_noise"
    duration_seconds : int
        Length of recording in seconds
    sample_rate : int
        Sample rate in Hz (default: 16000)
    channels : int
        Number of audio channels (default: 1 for mono)
    output_dir : str
        Directory to save recordings (default: 'sessions')
    
    Returns:
    --------
    str
        Path to the saved file
    """
    
    # Validate session type
    valid_types = ["glucose_alarm", "no_glucose_alarm"]
    if session_type not in valid_types:
        raise ValueError(f"session_type must be one of {valid_types}")
    
    # Validate background noise
    valid_noise = ["background_noise", "no_background_noise", "alarms"]
    if background_noise not in valid_noise:
        raise ValueError(f"background_noise must be one of {valid_noise}")
    
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Generate timestamp
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    
    # Generate filename: session_<timestamp>__<session_type>__<background_noise>.wav
    filename = f"session_{timestamp}__{session_type}__{background_noise}.wav"
    filepath = os.path.join(output_dir, filename)
    
    # Display recording info
    print("=" * 70)
    print("AUDIO SESSION RECORDING")
    print("=" * 70)
    print(f"Session Type: {session_type}")
    print(f"Background Noise: {background_noise}")
    print(f"Duration: {duration_seconds} seconds")
    print(f"Sample Rate: {sample_rate} Hz")
    print(f"Channels: {channels} (mono)")
    print(f"Format: 16-bit PCM WAV")
    print(f"Output: {filepath}")
    print("=" * 70)
    print(f"\n🎤 Recording will start in 3 seconds...")
    
    # Countdown
    import time
    for i in range(3, 0, -1):
        print(f"   {i}...")
        time.sleep(1)
    
    print(f"\n🔴 RECORDING NOW ({duration_seconds} seconds)\n")
    
    # Record audio
    recording = sd.rec(
        int(duration_seconds * sample_rate),
        samplerate=sample_rate,
        channels=channels,
        dtype='int16'
    )
    
    # Wait for recording to complete
    sd.wait()
    
    print("\n✓ Recording complete!")
    
    # Save as WAV file
    wavfile.write(filepath, sample_rate, recording)
    
    # Get file size
    file_size_mb = os.path.getsize(filepath) / (1024 * 1024)
    
    # Display summary
    print("\n" + "=" * 70)
    print("SESSION SAVED")
    print("=" * 70)
    print(f"File: {filepath}")
    print(f"Size: {file_size_mb:.2f} MB")
    print(f"Samples: {len(recording):,}")
    print("=" * 70)
    
    return filepath
```

## Record Session

**Run this cell to start recording with the configured parameters above.**

```python
# Record session with configured parameters
filepath = record_session(
    session_type=SESSION_TYPE,
    background_noise=BACKGROUND_NOISE,
    duration_seconds=DURATION_SECONDS
)
```
